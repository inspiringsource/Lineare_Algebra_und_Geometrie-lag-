## Aufgabe 10

###  a) 

A sei eine 2x10-Matrix mit

$$a_{ij} = \left\{\begin{array}{cc}i&\text{für }i=1\\i^2&\text{für }i=2\end{array}\right.$$

```python
A = np.zeros((2,10))
for z in range(0,2):
    for s in range(0,10):
        if z==0:
            A[z,s] = z+1
        else:
            A[z,s] = (z+1)**2
print(A)
```

### b) 

B sei eine 6x6-Matrix mit

$$b_{ij} = \left\{\begin{array}{cc}0&\text{für }i=j\\i+j&\text{für }i<j\\i-j&\text{für }i>j\end{array}\right.$$

```python
B = np.zeros((6,6))
for z in range(0,6):
    for s in range(0,6):
        if z==s:
            B[z,s] = 0
        elif z<s:
            B[z,s] = (z+1)+(s+1)
        else:
            B[z,s] = (z+1)-(s+1)
print(B)
```

###  c) 

C sei eine 5x5-Matrix mit

$$c_{ij} = \left\{\begin{array}{cc}0&\text{für }i+j<6\\1&\text{für }i+j=6\\2&\text{für }i+j>6\end{array}\right.$$

```python
C = np.zeros((5,5))
for z in range(0,5):
    for s in range(0,5):
        if z+s<4:
            C[z,s] = 0
        elif z+s==4:
            C[z,s] = 1
        else:
            C[z,s] = 2
print(C)
```
