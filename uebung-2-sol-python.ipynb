## Aufgabe 12

### a) elementare Zeilenumformungen auf Koeffizientenmatrix

```python
def skalare_multiplikation( M, zeile, skalar ):
    """
    Multipliziert die Zeile <zeile> der Matrix M mit dem 
    Wert <skalar> und gibt das Resultat zurück.
    """
    
    M = M.copy()
    M[zeile,:] = M[zeile,:] * skalar
    
    return M
```

```python
def vertausche( M, zeile1, zeile2 ):
    """
    Vertauscht die Zeile <zeile1> der Matrix M mit der 
    Zeile <zeile2> und gibt das Resultat zurück.
    """

    M = M.copy()
    tmp = M[zeile1,:].copy()
    M[zeile1,:] = M[zeile2,:]
    M[zeile2,:] = tmp
    
    return M
```

```python
def addiere( M, zeile1, zeile2, k ):
    """
    Addiert <k> mal die Zeile <zeile2> zur Zeile <zeile1> dazu und gibt das Resultat zurück.
    """
    
    M = M.copy()
    M[zeile1,:] += k*M[zeile2,:]

    return M
```

```python
C = np.array([[0,5,5,5],[2,3,1,5],[3,2,-1,6],[1,0,-1,1]])
print(C)
```

```python
skalare_multiplikation( C, 1, 2 )
```

```python
vertausche( C, 2, 3 )
```

```python
addiere( C, 0, 2, 3 )
```

### b) Gauss-Algorithmus 'von Hand'

Schritt 1:

```python
C = vertausche( C, 0, 3 )
print(C)
```

Schritt 2:

```python
C = addiere( C, 1, 0, -2 )
C = addiere( C, 2, 0, -3 )
print(C)
```

Schritt 3:

```python
C = skalare_multiplikation( C, 1, 1/3 )
C = skalare_multiplikation( C, 3, 1/5 )
print( C )
```

Schritt 4:

```python
C = addiere( C, 2, 1, -2 )
C = addiere( C, 3, 1, -1 )
print( C )
```

## Aufgabe 13

```python
import matplotlib.pyplot as plt
```

### a)

$$
\left|
\begin{align}
x - y & = 1 \\
x + y & = -1
\end{align}
\right|
$$

```python
x = np.linspace( -5, 5, 100 )
y1 = x - 1
y2 = -x - 1

plt.figure( figsize=(7,7) )
plt.plot( x, y1, label="y = x - 1" )
plt.plot( x, y2, label="y = -x - 1" )
plt.xlabel("x")
plt.ylabel("y")
plt.grid( linestyle='--' )
plt.gca().set_aspect("equal")
plt.legend()
```

### b)

$$
\left|
\begin{align}
2x & = 1 \\
x - 3y & = 0
\end{align}
\right|
$$

```python
x = np.linspace( -10, 10, 100 )
x1 = 0.5
y2 = x/3

plt.figure( figsize=(7,7) )
plt.axvline( x1, label="x = 1/2" )
plt.plot( x, y2, label="y = x/3", c="orange" )
plt.xlabel("x")
plt.ylabel("y")
plt.grid( linestyle='--' )
plt.gca().set_aspect("equal")
plt.legend()
```
