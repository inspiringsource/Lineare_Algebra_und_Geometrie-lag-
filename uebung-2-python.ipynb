## Aufgabe 5

Bestimmen Sie die Werte von $k$ derart, dass die folgenden linearen Gleichungssysteme

- eine **eindeutige Lösung**,  
- **unendlich viele Lösungen** bzw.  
- **keine Lösung**  

haben.

### (a)

$$
\begin{cases}
-kx + y = 0 \\
x - ky = 0
\end{cases}
$$

### (b)

$$
\begin{cases}
kx + y + z = 1 \\
x + ky + z = 1 \\
x + y + kz = 1
\end{cases}
$$

### (c)

$$
\begin{cases}
x - 3z = -3 \\
2x + ky - z = -2 \\
x + 2y + kz = 1
\end{cases}
$$

### (d)

Für die Variablen $a, b, c$ und $d$:

$$
\begin{cases}
(2 + k)a = 0 \\
(2 - k)b = 0 \\
a - 2b - kc - d = 0 \\
2a - 4b + c - kd = k
\end{cases}
$$

**Anmerkung zur Aufgabe:**  
Die Gleichungen von linearen Gleichungssystemen formalisieren Bedingungen einer Anwendung. Manchmal hat man bei den Bedingungen noch einen Spielraum (den Parameter $k$), den man in Abhängigkeit der Anzahl der Lösungen wählt.

```python
import sympy as sp

k = sp.symbols('k')

A = sp.Matrix([
    [-k, 1],
    [1, -k]
])

b = sp.Matrix([0,0])

print(A)
```

```python
# Aufgabe 5 (a)
print(A.det())
```

```python
# Aufgabe 5 (b)
k = sp.symbols('k')

# coefficient matrix
A = sp.Matrix([
    [k, 1, 1],
    [1, k, 1],
    [1, 1, k]
])

# right side vector
b = sp.Matrix([1, 1, 1])

# determinant
detA = A.det()
print("det(A) =", detA)

# solve system
x, y, z = sp.symbols('x y z')
solution = sp.solve(A * sp.Matrix([x, y, z]) - b, (x, y, z))

print("solution =", solution)
```

## Aufgabe 6

Drei Punkte der Ebene, die **nicht** auf einer Geraden liegen, bestimmen einen Kreis.  
Die allgemeine Gleichung eines Kreises wird in der folgenden Form gegeben:

$$
a(x^2 + y^2) + b x + c y = 1
$$

Bestimmen Sie die Kreisleichung des Kreises, der durch die drei Punkte

- $A(1 \mid 1)$
- $B(4 \mid 12)$
- $C(3 \mid -3)$

verläuft.

**Hinweis:**  
Setzen Sie die drei Punkte nacheinander in die Gleichung ein und lösen Sie das entstehende lineare Gleichungssystem für $a$, $b$ und $c$.

```python
import sympy as sp

# unknowns
a, b, c = sp.symbols('a b c')

# points
points = [(1,1), (4,12), (3,-3)]

# equations from the circle formula
eq1 = a*(points[0][0]**2 + points[0][1]**2) + b*points[0][0] + c*points[0][1] - 1
eq2 = a*(points[1][0]**2 + points[1][1]**2) + b*points[1][0] + c*points[1][1] - 1
eq3 = a*(points[2][0]**2 + points[2][1]**2) + b*points[2][0] + c*points[2][1] - 1

# solve the linear system
solution = sp.solve([eq1, eq2, eq3], (a, b, c))

print("a, b, c =", solution)

# build the final circle equation
x, y = sp.symbols('x y')
circle_eq = solution[a]*(x**2 + y**2) + solution[b]*x + solution[c]*y

print("Circle equation:")
sp.pprint(circle_eq)
```

## Aufgabe 7

Drei Zahnräder eines Getriebes haben zusammen 80 Zähne. Bei zehn Umdrehungen des ersten
Rades drehen sich das zweite 18 - und das dritte 45-mal. Wieviele Zähne hat jedes Rad?

## ⋆ Task 7 (English translation)

Three gears of a transmission have a total of 80 teeth. When the first gear makes ten revolutions, the second gear turns 18 times and the third 45 times. How many teeth does each gear have?

```python
# our variables
z1, z2, z3 = sp.symbols('z1 z2 z3')

# equations
eq1 = z1 + z2 + z3 - 80
eq2 = 10*z1 - 18*z2
eq3 = 18*z2 - 45*z3

# solve system
solution = sp.solve([eq1, eq2, eq3], (z1, z2, z3))

print(solution)
```
