## Teil 1: Datentypen

Das grundlegende Fundament jeder Programmiersprache ist die Fähigkeit zur Speicherung von Information im Arbeitsspeicher. Dazu werden **Variablen** benutzt, deren Namen eigentlich nur dazu dienen, dass wir uns nicht die genaue Speicheradresse im RAM merken müssen, sondern nur eine Zeichenkette unserer Wahl. Hier ein Beispiel:
